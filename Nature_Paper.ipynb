### Frechet Distance Feature

```python
# Calculate Frechet distance function
def calculate_frechet_distance(voltage, time):
    # Ensure both curves have the same length
    min_len = min(len(voltage), len(time))
    voltage = voltage[:min_len]
    time = time[:min_len]
    
    # Compute the Frechet distance using a brute-force approach
    frechet_dist = np.inf
    for i in range(len(voltage)):
        for j in range(len(time)):
            dist = np.sqrt((voltage[i] - time[j]) ** 2)  # Euclidean distance
            frechet_dist = min(frechet_dist, dist)
    
    return frechet_dist

# Group data by 'Cycle_Index'
grouped_data = cycle_data_pro[cycle_data_pro['Res1'] > 0].groupby('Cycle_Index')

# Initialize lists to store Frechet distance values
frechet_distance_values = []

# Iterate over each cycle
for cycle, data_group in grouped_data:
    # Extract 'Voltage' and 'Time' data for the current cycle
    voltage = data_group['Voltage'].values
    time = data_group['Time'].values
    
    # Calculate Frechet distance
    distance = calculate_frechet_distance(voltage, time)
    frechet_distance_values.append(distance)

# Create a DataFrame to store results
frechet_distance_CC = pd.DataFrame({'Cycle_Index': grouped_data.groups.keys(),
                                    'Frechet_Distance_CC': frechet_distance_values})
```

### Hausdorff Distance Feature

```python
# Calculate Hausdorff distance function
def calculate_hausdorff_distance(voltage, time):
    # Compute pairwise distances between points in 'voltage' and 'time' arrays
    distances_vt = np.sqrt((np.subtract.outer(voltage, time))**2)
    distances_tv = np.sqrt((np.subtract.outer(time, voltage))**2)
    
    # Find the maximum distance from a point in 'voltage' to the nearest point in 'time'
    hausdorff_distance = np.max(np.min(distances_vt, axis=1))
    
    # Find the maximum distance from a point in 'time' to the nearest point in 'voltage'
    hausdorff_distance = max(hausdorff_distance, np.max(np.min(distances_tv, axis=1)))
    
    return hausdorff_distance

# Group data by 'Cycle_Index'
grouped_data = cycle_data_pro[cycle_data_pro['Res1'] > 0].groupby('Cycle_Index')

# Initialize lists to store Hausdorff distance values
hausdorff_distance_values = []

# Iterate over each cycle
for cycle, data_group in grouped_data:
    # Extract 'Voltage' and 'Time' data for the current cycle
    voltage = data_group['Voltage'].values
    time = data_group['Time'].values
    
    # Calculate Hausdorff distance
    distance = calculate_hausdorff_distance(voltage, time)
    hausdorff_distance_values.append(distance)

# Create a DataFrame to store results
hausdorff_distance_CC = pd.DataFrame({'Cycle_Index': grouped_data.groups.keys(),
                                      'Hausdorff_Distance_CC': hausdorff_distance_values})
```

### Shanon Entropy Feature

```python
# Calculate entropy function
def calculate_entropy(data):
    # Compute the probability distribution of the data
    unique_values, counts = np.unique(data, return_counts=True)
    probabilities = counts / len(data)
    
    # Apply the entropy formula
    entropy = -np.sum(probabilities * np.log2(probabilities))
    
    return entropy

# Group data by 'Cycle_Index'
grouped_data = cycle_data_pro[cycle_data_pro['Res1'] > 0].groupby('Cycle_Index')

# Initialize lists to store entropy values
entropy_values = []

# Iterate over each cycle
for cycle, data_group in grouped_data:
    # Extract 'Voltage' and 'Time' data for the current cycle
    voltage = data_group['Voltage'].values
    time = data_group['Time'].values
    
    # Calculate entropy for 'Voltage' and 'Time' separately
    voltage_entropy = calculate_entropy(voltage)
    time_entropy = calculate_entropy(time)
    
    # Combine the entropy values if needed, for example, by taking the average
    combined_entropy = (voltage_entropy + time_entropy) / 2
    
    entropy_values.append(combined_entropy)

# Create a DataFrame to store results
shannon_entropy_CC = pd.DataFrame({'Cycle_Index': grouped_data.groups.keys(),
                           'Shannon_Entropy_CC': entropy_values})
```

### Entropy Feature

```python
# Calculate entropy function using the given formula
def calculate_curve_entropy(L, D, N):
    EC = np.log2(2 * L / D) / np.log2(N - 1)
    return EC

# Group data by 'Cycle_Index'
grouped_data = cycle_data_pro[cycle_data_pro['Res1'] > 0].groupby('Cycle_Index')

# Initialize lists to store entropy values
curve_entropy_values = []

# Iterate over each cycle
for cycle, data_group in grouped_data:
    # Extract 'Voltage' and 'Time' data for the current cycle
    voltage = data_group['Voltage'].values
    time = data_group['Time'].values
    
    # Compute the length of the plane curve (L)
    L = np.sum(np.abs(np.diff(voltage)))
    
    # Compute the diameter of the smallest hypersphere covering the curve (D)
    D = np.max(voltage) - np.min(voltage)
    
    # Compute the number of segments approximating the line (N - 1)
    N = len(voltage) - 1
    
    # Calculate entropy using the formula
    curve_entropy = calculate_curve_entropy(L, D, N)
    
    curve_entropy_values.append(curve_entropy)

# Create a DataFrame to store results
entropy_CC = pd.DataFrame({'Cycle_Index': grouped_data.groups.keys(),
                           'Entropy_CC': curve_entropy_values})
```
