```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Features.csv")
```

```python
df
```

```python
# Loop through each column in the DataFrame
for column in df.columns:
    if column != 'Cycle_Index':  # Skip cycle index column
        # Bivariate analysis
        sns.scatterplot(x='Cycle_Index', y=column, data=df)
        plt.title(f"Bivariate Analysis: {column} vs. Cycle Index")
        plt.xlabel("Cycle_Index")
        plt.ylabel(column)
        plt.show()
```
